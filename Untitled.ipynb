```python


import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Попытка использовать skimage для SSIM/MSE
try:
    from skimage.metrics import structural_similarity as ssim_sk, mean_squared_error as mse_sk
    HAS_SKIMAGE = True
except Exception as e:
    HAS_SKIMAGE = False
    print(" skimage не найден")
    
# ----- Fallback-функции на NumPy/OpenCV -----
def mse_np(img1, img2):
    img1 = img1.astype(np.float64)
    img2 = img2.astype(np.float64)
    return np.mean((img1 - img2) ** 2)

def ssim_np(img1, img2, C1=(0.01*255)**2, C2=(0.03*255)**2, win_size=11, sigma=1.5):
    # обе картинки должны быть одноканальные uint8
    if img1.ndim != 2 or img2.ndim != 2:
        raise ValueError("SSIM fallback ожидает одноканальные (grayscale) изображения.")
    img1 = img1.astype(np.float64)
    img2 = img2.astype(np.float64)
    gauss = cv2.getGaussianKernel(win_size, sigma)
    window = gauss @ gauss.T
    mu1 = cv2.filter2D(img1, -1, window)
    mu2 = cv2.filter2D(img2, -1, window)
    mu1_sq = mu1 * mu1
    mu2_sq = mu2 * mu2
    mu1_mu2 = mu1 * mu2
    sigma1_sq = cv2.filter2D(img1*img1, -1, window) - mu1_sq
    sigma2_sq = cv2.filter2D(img2*img2, -1, window) - mu2_sq
    sigma12   = cv2.filter2D(img1*img2, -1, window) - mu1_mu2
    ssim_map = ((2*mu1_mu2 + C1) * (2*sigma12 + C2)) / ((mu1_sq + mu2_sq + C1) * (sigma1_sq + sigma2_sq + C2))
    return float(ssim_map.mean()), ssim_map

def calc_mse_ssim(img_ref, img_test):
    if HAS_SKIMAGE:
        m = mse_sk(img_ref, img_test)
        s, _ = ssim_sk(img_ref, img_test, full=True, data_range=255)
        return float(m), float(s)
    else:
        m = mse_np(img_ref, img_test)
        s, _ = ssim_np(img_ref, img_test)
        return float(m), float(s)

# ----- Загрузка серого изображения -----
def load_gray(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"Файл '{path}' не найден в рабочей папке {os.getcwd()}")
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)  # сразу в gray
    if img is None:
        raise ValueError(f"Не удалось загрузить '{path}'. Проверь расширение/кодировку.")
    return img

# ----- Гамма-коррекция -----
def gamma_correct(img_gray_u8, gamma: float):
    # вход: uint8 [0..255]
    img_f = img_gray_u8.astype(np.float32) / 255.0
    out = np.power(img_f, gamma)
    out = np.clip(out * 255.0, 0, 255).astype(np.uint8)
    return out

# ----- Статистическая коррекция: матчинг среднего и СКО -----
def match_mean_std(src_u8, ref_u8):
    # переводим к float, считаем статистики, выравниваем mean/std, клип и обратно в uint8
    s = src_u8.astype(np.float32)
    r = ref_u8.astype(np.float32)
    mu_s, std_s = float(s.mean()), float(s.std())
    mu_r, std_r = float(r.mean()), float(r.std())
    if std_s < 1e-6:
        std_s = 1e-6
    out = (s - mu_s) * (std_r / std_s) + mu_r
    return np.clip(out, 0, 255).astype(np.uint8)
```

```python
#1
src_path = 'sar_1_gray.jpg'    # исходное серое изображение
eq_path  = 'eq_gray.jpg'       # эталон для статистической коррекции 
image_gray = load_gray(src_path)
print("Размер исходного изображения:", image_gray.shape)

plt.figure(figsize=(5,5))
plt.imshow(image_gray, cmap='gray')
plt.title('Исходное изображение (gray)')
plt.axis('off')
plt.show()
```

```python
#2
plt.figure(figsize=(6,4))
plt.hist(image_gray.ravel(), bins=256, range=(0,255))
plt.title('Гистограмма яркости (0..255)')
plt.xlabel('Яркость')
plt.ylabel('Количество пикселей')
plt.show()

print("Статистика исходного:", f"mean={image_gray.mean():.2f}", f"std={image_gray.std():.2f}")
```

```python
#3
gamma_lt = 0.6  # < 1 — делает изображение светлее в тенях
gamma_gt = 1.8  # > 1 — затемняет тени, усиливает контраст в светлых

img_gamma_lt = gamma_correct(image_gray, gamma_lt)
img_gamma_gt = gamma_correct(image_gray, gamma_gt)

# Визуализация
fig, ax = plt.subplots(1,3, figsize=(14,4))
ax[0].imshow(image_gray, cmap='gray'); ax[0].set_title('Оригинал'); ax[0].axis('off')
ax[1].imshow(img_gamma_lt, cmap='gray'); ax[1].set_title(f'Gamma {gamma_lt}'); ax[1].axis('off')
ax[2].imshow(img_gamma_gt, cmap='gray'); ax[2].set_title(f'Gamma {gamma_gt}'); ax[2].axis('off')
plt.show()

# Метрики vs исходного
m_lt, s_lt = calc_mse_ssim(image_gray, img_gamma_lt)
m_gt, s_gt = calc_mse_ssim(image_gray, img_gamma_gt)
print(f"[Gamma {gamma_lt}] MSE={m_lt:.4f}, SSIM={s_lt:.4f}")
print(f"[Gamma {gamma_gt}] MSE={m_gt:.4f}, SSIM={s_gt:.4f}")
```

```python
#4
diff_lt = cv2.absdiff(image_gray, img_gamma_lt)
diff_gt = cv2.absdiff(image_gray, img_gamma_gt)

fig, ax = plt.subplots(1,2, figsize=(10,4))
ax[0].imshow(diff_lt, cmap='gray'); ax[0].set_title('|Ориг - Gamma<1|'); ax[0].axis('off')
ax[1].imshow(diff_gt, cmap='gray'); ax[1].set_title('|Ориг - Gamma>1|'); ax[1].axis('off')
plt.show()
```

```python
#6
# Нормализация для стабильности 
img = image_gray

# Фиксированные глобальные пороги
thresholds = [64, 128, 192]
bin_fixed = [(t, (img > t).astype(np.uint8)*255) for t in thresholds]

# Otsu
_, bin_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Адаптивные пороги (mean и gaussian)
bin_adapt_mean = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                       cv2.THRESH_BINARY, blockSize=31, C=5)
bin_adapt_gauss = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                        cv2.THRESH_BINARY, blockSize=31, C=5)

# Печать результатов (как минимум показать изображения)
cols = 3
rows = 1 + int(np.ceil((len(bin_fixed)+3) / cols))
plt.figure(figsize=(14, 4*rows))

idx = 1
plt.subplot(rows, cols, idx); idx+=1
plt.imshow(img, cmap='gray'); plt.title('Оригинал'); plt.axis('off')

for t, bimg in bin_fixed:
    plt.subplot(rows, cols, idx); idx+=1
    plt.imshow(bimg, cmap='gray'); plt.title(f'Глоб. порог {t}'); plt.axis('off')

plt.subplot(rows, cols, idx); idx+=1
plt.imshow(bin_otsu, cmap='gray'); plt.title('Otsu'); plt.axis('off')

plt.subplot(rows, cols, idx); idx+=1
plt.imshow(bin_adapt_mean, cmap='gray'); plt.title('Adaptive Mean (31, C=5)'); plt.axis('off')

plt.subplot(rows, cols, idx); idx+=1
plt.imshow(bin_adapt_gauss, cmap='gray'); plt.title('Adaptive Gaussian (31, C=5)'); plt.axis('off')

plt.tight_layout()
plt.show()

# "Распечатать результат" в тексте — опишем базовые метрики/счётчики
def foreground_ratio(binary_u8):
    return float(np.mean(binary_u8 > 0))

print("— Результаты пороговой фильтрации —")
for t, bimg in bin_fixed:
    print(f"Глоб. порог {t}: доля белых пикселей = {foreground_ratio(bimg):.3f}")
print(f"Otsu: доля белых = {foreground_ratio(bin_otsu):.3f}")
print(f"Adaptive Mean: доля белых = {foreground_ratio(bin_adapt_mean):.3f}")
print(f"Adaptive Gaussian: доля белых = {foreground_ratio(bin_adapt_gauss):.3f}")
```

```python
#5
# eq_gray уже создан выше
img_stat_corr = match_mean_std(image_gray, eq_gray)

m_stat, s_stat = calc_mse_ssim(image_gray, img_stat_corr)

fig, ax = plt.subplots(1,3, figsize=(14,4))
ax[0].imshow(image_gray, cmap='gray'); ax[0].set_title('Оригинал'); ax[0].axis('off')
ax[1].imshow(eq_gray, cmap='gray'); ax[1].set_title('Эталон eq_gray'); ax[1].axis('off')
ax[2].imshow(img_stat_corr, cmap='gray'); ax[2].set_title('Стат. коррекция (mean/std)'); ax[2].axis('off')
plt.show()

print(f"[Stat-correction vs Original] MSE={m_stat:.4f}, SSIM={s_stat:.4f}")
print("Статистика после коррекции:", f"mean={img_stat_corr.mean():.2f}", f"std={img_stat_corr.std():.2f}")
print("Целевая статистика (eq_gray):", f"mean={eq_gray.mean():.2f}", f"std={eq_gray.std():.2f}")

```
